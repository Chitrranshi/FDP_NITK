```python
link ="https://economictimes.indiatimes.com/news/narendra-modi"
root ="https://economictimes.indiatimes.com"

```

```python
#Function to write the article content to a file
def write_to_file(title, f_art):
    #file name
    f_name = os.path.join(output_dir,title+".txt")
    #Writting to file
    document = open(f_name, "w")
    document.write(f_art)
    document.close()

#Function to Create CSV File for all the records
def create_csv(link, date, title):
    #creating CSV file
    document = open("ET.csv", "a")
    document.write("{}, {}, {} \n".format(date, title, link))
    document.close()


#function for creating Article.txt file
def create_article(news_link,title):
    print("getting article")
    headers={'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36'}
    req = requests.get(news_link, headers)
    
    webpage = req.text       
    soup = BeautifulSoup(webpage, 'html5lib')
    #Scrapping date and cleaning it
    date = soup.find('time').get_text()
    date = date.replace(",","")
    date = date.strip()
    date = date.replace("\n"," ")
    #calling Article function from newspaper then downloading and parsing it
    toi_article = Article(link, language="en")
    toi_article.download()
    toi_article.parse()
    #Formating content for article.txt
    f_art = f"{date}\n\n\n{title}\n\n\n{toi_article.text}\n\n\n{link}"
    #Calling write_to_file to write it to file
    write_to_file(title,f_art)
    print("written to csv")
    create_csv(link, date, title)


```

```python
def data_Extration(search_link):
    print(search_link)
    headers={'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36'}
    req = requests.get(search_link,headers )
    print(req)    
    webpage = req.text
    #print(webpage)
    soup = BeautifulSoup(webpage, 'html5lib')
    #print(soup)
    try:
        #Finding all div in class authorListing
        for item in soup.find_all('div', attrs={'class': 'eachStory'}):
            #Finding all the a tags
            url = item.find('a', href=True)['href']
            #print(url)
            news_link = root + url
            title = item.find('h3').get_text()
            title = title.replace(",","")
            title = title.replace("\n"," ")
            title = title.replace("\\","-")
            title = title.replace("/","-")
            title = title.strip()
            #print(title)
            #calling function to create CSV and article.txt
            print(news_link)
            create_article(news_link,title)
    except:
        print("Exception in find ")





```

```python
#Standarlization: to re-scale the feature values to their unit variance.

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()


X_scaled = scaler.fit_transform(X)
```

```python
# Import the classifiers

from sklearn.linear_model import LogisticRegression

from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier



from sklearn.metrics import roc_curve, roc_auc_score


# Instantiate the classfiers and make a list

classifiers = [LogisticRegression(random_state=1234), 
               GaussianNB(), 
               KNeighborsClassifier(), 
               DecisionTreeClassifier(random_state=1234),
               RandomForestClassifier(random_state=1234)]

# Define a result table as a DataFrame

result_table = pd.DataFrame(columns=['classifiers', 'fpr','tpr','auc'])

print(result_table)
```

```python
# Train the models and record the results

for cls in classifiers:
    model = cls.fit(X_train, y_train)
    yproba = model.predict_proba(X_test)[::,1]
    
    fpr, tpr, _ = roc_curve(y_test,  yproba)
    auc = roc_auc_score(y_test, yproba)
    
    result_table = result_table.append({'classifiers':cls.__class__.__name__,
                                        'fpr':fpr, 
                                        'tpr':tpr, 
                                        'auc':auc}, ignore_index=True)

# Set name of the classifiers as index labels
result_table.set_index('classifiers', inplace=True)
print(result_table)
```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt


fig = plt.figure(figsize=(8,6))


for i in result_table.index:
    plt.plot(result_table.loc[i]['fpr'], 
             result_table.loc[i]['tpr'], 
             label="{}, AUC={:.3f}".format(i, result_table.loc[i]['auc']))
    
plt.plot([0,1], [0,1], color='orange', linestyle='--')

plt.xticks(np.arange(0.0, 1, step=0.1))
plt.xlabel("Flase Positive Rate", fontsize=15)

plt.yticks(np.arange(0.0, 1, step=0.1))
plt.ylabel("True Positive Rate", fontsize=15)

plt.title('ROC Curve Analysis', fontweight='bold', fontsize=15)
plt.legend(prop={'size':13}, loc='lower right')

plt.show()
```
