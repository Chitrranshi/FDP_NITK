#WORKING WITH A DATASET

```python
import pandas as pd
df = pd.read_csv('/content/sample_data/Salary.csv')
df.head(10)  #printing first 10 values
top_10 = df['Salary'].sort_values(ascending = False).head(10)
top_10
top_10 = top_10.sort_index()
type(top_10)
top_10 = top_10.to_frame()
```

#MULTIPLE BAR CHART

```python
import numpy as np
x = [1,2,3,4,5]
experience = [1,2,3,4,5]
salary_Percentage = [0.3 , 0.4 , 0.5 , 0.6 , 0.7]
width = 0.4
bar1 = np.arange(len(x))
bar2 = [i+width for i in bar1]

plt.bar(bar1 , experience , width , label = 'Experience' , edgecolor = 'k' , color = 'navy')
plt.bar(bar2 , salary_Percentage , width , label = 'Percentage of salary' , edgecolor = 'k' , color = 'slategrey')
plt.xlabel('Rank')
plt.ylabel('Experience & Salary')
plt.title('Top 5 Employee - Salary and Experience')
plt.legend(edgecolor = 'k')
plt.xticks(bar1 , x)
```

#pie chart

```python
df.info()
exp_yrs = df['YearsExperience']
x1 = [i for i in exp_yrs if 0<i<5]
x1 = len(x1)
x2 = [i for i in exp_yrs if 5<i<10]
x2 = len(x2)
x3 = [i for i in exp_yrs if 10<i<15]
x3 = len(x3)
my_pie = [x1,x2,x3]
labels = ['0-5(yrs)' , '5-10(yrs)' , '10-15(yrs)']
colors = ['lightsteelblue' , 'cornflowerblue' , 'royalblue']
explode = [0.1 , 0 , 0]
plt.title('Proportion of Years of Experience')
plt.tight_layout()
plt.pie(my_pie , labels = labels , colors = colors ,  wedgeprops = {'edgecolor' : 'black'} , 
        explode = explode , shadow = True , startangle= 45 , autopct = '%1.1f%%')
plt.show()
```

#STACK PLOTS

```python
year = [2017 , 2018 , 2019 , 2020]
salary_below_8 = [17761 , 29343 , 43515 , 66105]
salary_above_8 = [37791 , 50344 , 63575, 86140 ]
labels_stack = ['below 8 years of experience' , 'above 8 years of experience'] 
colors = ['lightseagreen', 'mediumturquoise']
plt.stackplot(year , salary_below_8 , salary_above_8 , labels = labels_stack , colors = colors)
plt.title('Salary over years based on Experience')
plt.xlabel('Year')
plt.ylabel('Salary')
plt.legend(loc = (0.05 , 0.7))
```

#HISTOGRAM

```python
exp = df['YearsExperience']
mean = exp.mean()

exp
bins = [0,2,4,6,8,10,12,14]
plt.hist(exp, bins = bins , edgecolor = 'black' , color = 'thistle')
plt.axvline(mean , color = 'darkgoldenrod' , label = 'median line')
plt.title('Number of people with corresponding years of Experience')
plt.xlabel('Experience')
plt.ylabel('Count of employees')
plt.legend(edgecolor = 'k')
```

#SCATTERPLOT

```python
plt.style.use('seaborn')

x = df['YearsExperience']
y = df['Salary']
plt.scatter(x,y , s = 50 , edgecolor = 'k' ,  linewidth = 1 , alpha = 0.75)
plt.title('Experience Vs Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

#TIMESERIES

```python
data = pd.read_csv('/content/sample_data/SalesData.csv')
data['date'] = pd.to_datetime(data['date'])
data = data.resample('YS' , on = "date").sum() 
data.info()
dates = data.index
beer_price = data['beer']

plt.plot_date(dates , beer_price , linestyle = 'solid' , MarkerFacecolor = 'darkblue' , color = 'k' , marker = '*' )
plt.title('Time Series plot for beer price over years')
plt.xlabel('Year')
plt.ylabel('Price')
plt.show()
```
